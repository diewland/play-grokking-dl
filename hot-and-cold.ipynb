```python
weight = 0.5
input = 0.5
goal_pred = 0.8

step_amount = 0.001

for iteration in range(1101):
    pred = input * weight
    err  = (pred-goal_pred)**2
    print "Error: %s Predition: %s" % ( err, pred )
    
    up_pred = input * (weight+step_amount)
    up_err  = (up_pred-goal_pred)**2
    
    down_pred = input * (weight-step_amount)
    down_err  = (down_pred-goal_pred)**2
    
    if(down_err < up_err):
        weight -= step_amount
    elif(down_err > up_err):
        weight += step_amount
```

```python
weight = 0.5
goal_pred = 0.8
input = 0.5

for iteration in range(20):
    pred = input * weight
    err  = (pred-goal_pred)**2
    direction_and_amount = (pred-goal_pred) * input
    weight -= direction_and_amount
    
    print "Error %s Prediction %s" % (err, pred)
```
